```python
import pandas as pd

# Replace 'your_file.csv' with the actual filename
df = pd.read_csv('exp1_bmi.csv', encoding='ISO-8859-1')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Importing required libraries for descriptive statistics
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Descriptive Statistics - Central Tendency
# Calculate Mean, Median, Mode
mean_bmi = df['bmi'].mean()
median_bmi = df['bmi'].median()
mode_bmi = df['bmi'].mode()[0]

# Print the results for BMI
print(f'Mean BMI: {mean_bmi}')
print(f'Median BMI: {median_bmi}')
print(f'Mode BMI: {mode_bmi}')

# For other columns: Height, Weight, Age
mean_height = df['Height'].mean()
mean_weight = df['Weight'].mean()
mean_age = df['Age'].mean()

print(f'Mean Height: {mean_height}')
print(f'Mean Weight: {mean_weight}')
print(f'Mean Age: {mean_age}')
```

```python
# 2. Descriptive Statistics - Measures of Spread/Dispersion
# Variance and Standard Deviation
var_bmi = df['bmi'].var()
std_bmi = df['bmi'].std()

var_height = df['Height'].var()
std_height = df['Height'].std()

var_weight = df['Weight'].var()
std_weight = df['Weight'].std()

var_age = df['Age'].var()
std_age = df['Age'].std()

print(f'BMI Variance: {var_bmi}, BMI Standard Deviation: {std_bmi}')
print(f'Height Variance: {var_height}, Height Standard Deviation: {std_height}')
print(f'Weight Variance: {var_weight}, Weight Standard Deviation: {std_weight}')
print(f'Age Variance: {var_age}, Age Standard Deviation: {std_age}')

# Range (Max - Min) for each column
range_bmi = df['bmi'].max() - df['bmi'].min()
range_height = df['Height'].max() - df['Height'].min()
range_weight = df['Weight'].max() - df['Weight'].min()
range_age = df['Age'].max() - df['Age'].min()

print(f'BMI Range: {range_bmi}')
print(f'Height Range: {range_height}')
print(f'Weight Range: {range_weight}')
print(f'Age Range: {range_age}')

# Interquartile Range (IQR)
iqr_bmi = df['bmi'].quantile(0.75) - df['bmi'].quantile(0.25)
iqr_height = df['Height'].quantile(0.75) - df['Height'].quantile(0.25)
iqr_weight = df['Weight'].quantile(0.75) - df['Weight'].quantile(0.25)
iqr_age = df['Age'].quantile(0.75) - df['Age'].quantile(0.25)

print(f'BMI IQR: {iqr_bmi}')
print(f'Height IQR: {iqr_height}')
print(f'Weight IQR: {iqr_weight}')
print(f'Age IQR: {iqr_age}')
```

```python
# 3. Descriptive Statistics - Measures of Shape
# Skewness
skewness_bmi = df['bmi'].skew()
skewness_height = df['Height'].skew()
skewness_weight = df['Weight'].skew()
skewness_age = df['Age'].skew()

print(f'BMI Skewness: {skewness_bmi}')
print(f'Height Skewness: {skewness_height}')
print(f'Weight Skewness: {skewness_weight}')
print(f'Age Skewness: {skewness_age}')

# Kurtosis
kurtosis_bmi = df['bmi'].kurtosis()
kurtosis_height = df['Height'].kurtosis()
kurtosis_weight = df['Weight'].kurtosis()
kurtosis_age = df['Age'].kurtosis()

print(f'BMI Kurtosis: {kurtosis_bmi}')
print(f'Height Kurtosis: {kurtosis_height}')
print(f'Weight Kurtosis: {kurtosis_weight}')
print(f'Age Kurtosis: {kurtosis_age}')
```

```python
# 4. Boxplot for Visual Representation of Dispersion
# Boxplots for BMI, Height, Weight, Age
plt.figure(figsize=(10, 8))

# BMI Boxplot
plt.subplot(2, 2, 1)
sns.boxplot(y=df['bmi'])
plt.title('Boxplot of BMI')

# Height Boxplot
plt.subplot(2, 2, 2)
sns.boxplot(y=df['Height'])
plt.title('Boxplot of Height')

# Weight Boxplot
plt.subplot(2, 2, 3)
sns.boxplot(y=df['Weight'])
plt.title('Boxplot of Weight')

# Age Boxplot
plt.subplot(2, 2, 4)
sns.boxplot(y=df['Age'])
plt.title('Boxplot of Age')

plt.tight_layout()
plt.show()
```
