```python
from google.colab import files
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Upload the dataset
uploaded = files.upload()
```

```python
# Check the names of the uploaded files
uploaded.keys()
```

```python
# Load the uploaded file into a pandas DataFrame (Assuming the uploaded file is CSV)
file_name = list(uploaded.keys())[0]  # Get the name of the uploaded file
data = pd.read_excel(file_name)
```

```python
# Show the first few rows of the dataset to understand its structure
data.head()
```

```python
# Assuming the column name for rice bag weights is 'Rice_Bag_Weight'
# If the dataset contains a different column name, adjust accordingly.
rice_bag_weights = data['Rice_Bag_Weight'].dropna()  # Remove any NaN values
```

```python
# Step 2: Calculate sample statistics
sample_mean = np.mean(rice_bag_weights)
sample_std = np.std(rice_bag_weights, ddof=1)  # Sample standard deviation (ddof=1 for sample)
sample_size = len(rice_bag_weights)
```

```python
# Step 3: Calculate the standard error of the mean
std_error = sample_std / np.sqrt(sample_size)
```

```python
# Step 4: Perform One-Sample t-Test (using population mean of 24.5 as an example)
population_mean = 24.44621
t_statistic, p_value = stats.ttest_1samp(rice_bag_weights, population_mean)
```

```python
# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Sample Size: {sample_size}")
print(f"Standard Error: {std_error}")
print(f"t-Statistic: {t_statistic}")
print(f"p-Value: {p_value}")
```

```python
# Step 5: Interpretation of the t-Test
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference from the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference from the population mean.")
```

```python
# Step 6: Visualizing the distribution of rice bag weights
plt.figure(figsize=(10, 6))
sns.histplot(rice_bag_weights, kde=True, color='blue', bins=10)
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label=f'Population Mean ({population_mean})')
plt.title('Distribution of Rice Bag Weights')
plt.xlabel('Rice Bag Weight')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Step 7: Summary Statistics
print("\nSummary Statistics:")
print(f"Mean: {sample_mean}")
print(f"Median: {np.median(rice_bag_weights)}")

# Correct way to access the mode
mode_result = stats.mode(rice_bag_weights)
# print(f"Mode: {mode_result.mode[0]}")  # Accessing the first mode value

print(f"Standard Deviation: {sample_std}")
print(f"Variance: {np.var(rice_bag_weights)}")
print(f"Kurtosis: {stats.kurtosis(rice_bag_weights)}")
print(f"Skewness: {stats.skew(rice_bag_weights)}")
print(f"Range: {np.ptp(rice_bag_weights)}")
print(f"Minimum: {min(rice_bag_weights)}")
print(f"Maximum: {max(rice_bag_weights)}")
```
